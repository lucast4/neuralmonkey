```python
def df_cleanup(dfdat_var_events, SP):
    """ Clean up df_var..."""
    from pythonlib.tools.pandastools import applyFunctionToAllRows, aggregGeneral
    
    sn = SP.SN
    
    # just keep the r2
    dfdat_var_events = dfdat_var_events[dfdat_var_events["val_kind"]=="modulation_across_events"].reset_index(drop=True)
    
    # 1) group events into baseline, rule cue, samp, go(pre), stroke1 (during).
    if sn.Date=="221125": 
        mapper_simpleevents_to_events = {
            "baseline":["00_fixcue", "01_fixcue", "02_fix_touch"],
            "cue_post":["03_fix_touch", "04_samp"],
            "image_post":["05_samp"],
            "movement_pre":["06_first_raise"],
            "stroke1_during":["07_on_strokeidx_0", "08_on_strokeidx_0"]}
    elif sn.Date=="221020": 
        mapper_simpleevents_to_events = {
            "baseline":["00_fixcue"],
            "cue_post":["01_fixcue", "02_fix_touch", "03_samp"],
            "image_post":["04_samp"],
            "movement_pre":["05_first_raise"],
            "stroke1_during":["06_on_strokeidx_0", "07_on_strokeidx_0"]}
    elif sn.Date in ["221031", "221102", "221114"]: 
        mapper_simpleevents_to_events = {
            "baseline":["00_fixcue", "01_fixcue", "02_fix_touch"],
            "cue_post":["03_fix_touch", "04_samp"],
            "image_post":["05_samp"],
            "movement_pre":["06_first_raise"],
            "stroke1_during":["07_on_strokeidx_0", "08_on_strokeidx_0"]}
    else:
        assert False, "current hand coded for each day"

    # create inverse mapper
    mapper_events_to_simpleevents = {}
    for key, values in mapper_simpleevents_to_events.items():
        for ev in values:
            assert ev not in mapper_events_to_simpleevents
            if ev not in SP.Params["list_events_uniqnames"]:
                print(ev)
                print(SP.Params["list_events_uniqnames"])
                assert False, "failed sanity cehck. typo?"
            mapper_events_to_simpleevents[ev] = key

    # aggregate the events.
    def F(x):
        ev = x["event"]
        if ev in mapper_events_to_simpleevents.keys():
            return mapper_events_to_simpleevents[ev]
        else:
            return "IGNORE"
    dfdat_var_events = applyFunctionToAllRows(dfdat_var_events, F, "simpleevent")
    print(dfdat_var_events["simpleevent"].value_counts())
    
    # Create aggretaged version
    dfdat_var_events_SIMPLEEVENTS = aggregGeneral(dfdat_var_events, ["var", "chan", "simpleevent", "val_kind", "load_suffix"], ["val"], nonnumercols=["bregion"])
    # remove those with IGNORE events
    dfdat_var_events_SIMPLEEVENTS = dfdat_var_events_SIMPLEEVENTS[~(dfdat_var_events_SIMPLEEVENTS["simpleevent"]=="IGNORE")].reset_index(drop=True)

    # reassigne breain areas
    # aggregate regions
    # (any regions not included here will not be plotted)
    map_aggregion_to_region = {
        "M1":["M1_m", "M1_m"],
        "PMv_l":["PMv_l"],
        "PMd":["PMd_p", "PMd_a"],
        "PMd":["PMd_p", "PMd_a"],
    }

    
    # }
    def F(x):
        # map to specific subregion
        return sn.sitegetter_map_site_to_region(x["chan"], True)
    dfdat_var_events_SIMPLEEVENTS = applyFunctionToAllRows(dfdat_var_events_SIMPLEEVENTS, F, "simplebregion")

#     # only plot specific regions, simpler plots
#     REGIONS_KEEP = ["M1", "PMd", "SMA", "dlPFC", ]

    ##### Plots
    return dfdat_var_events, dfdat_var_events_SIMPLEEVENTS
```

```python
def _load_date(date, which_index_to_use_for_SP = 0):
    if False:
        # Stricter units:
        # - 90mV threshold (spikes)
        # - firing rate threshold
        list_suffix = [
            "rulesgoodfinal_all",
            "rulesgoodfinal_samebeh",
            "rulesgoodfinal_samefirststroke",
        ]
        assert False, "use conditions as in lenient units"
    else:
        # Lenient units
        # - 70mV
        # - (ignore fr constraint)
        basesuffix = "rulesgoodfinalmoreunits"
        listcondition = ["all", "samebeh", "samefirststroke"]
        list_sess = [0, 0, 0]
        note = "Simons site visit final analysis on rules"


    # save dir
    savedir = f"{SAVEDIR}/{date}"
    os.makedirs(savedir, exist_ok=True)
    from pythonlib.tools.expttools import writeDictToYaml

    savedict = {
        "basesuffix":basesuffix,
        "listcondition":listcondition,
        "list_sess":list_sess,
        "note":note
    }

    writeDictToYaml(savedict, f"{savedir}/extraction_params.yaml")


    LIST_RES = []
    LIST_SP = []
    for i, (cond, sess) in enumerate(zip(listcondition, list_sess)):

        suff = f"{basesuffix}_{cond}"
        print(suff)
        path = f"{DIR}/{date}_{suff}/sess_{sess}/scalar_plots/RES_ALL_CHANS.pkl"
        with open(path, "rb") as f:
            resallchans = pickle.load(f)

    #     path = f"{DIR}/{date}_{suff}/sess_{sess}/scalar_plots/SP.pkl"
    #     sp = pickle.load(path)

        LIST_RES.append(resallchans)
        if i==which_index_to_use_for_SP:
            # load a single SP for its methods
            path = f"{DIR}/{date}_{suff}/sess_{sess}/scalar_plots/SP.pkl"
            with open(path, "rb") as f:
                SP = pickle.load(f)
    #     LIST_SP.append(sp)


    # Combine all into a single dataframe
    RESALLALL = []
    for resallchan, suff in zip(LIST_RES, listcondition):
        for res in resallchan:
            res["load_suffix"] = suff
            RESALLALL.append(res)
    #         print(res)
    #         assert False

    # Compute modulation
    OUT = SP.modulation_compute_higher_stats(RESALLALL)

    # Post-process, and get simplified (aggregated) version

    dfdat = OUT["dfdat_var_events"]
    dfdat, dfdat_simple = df_cleanup(dfdat, SP)

    ##### Normalize, subtracting baseline
    # Plot after normalizing
    from pythonlib.tools.pandastools import aggregThenReassignToNewColumn, applyFunctionToAllRows

    def F(x):
        this = x[x["simpleevent"]=="baseline"]
        if len(this["val"])>1:
            assert len(this["val"])<4, "seems to omany. bug?"
            baseval = np.mean(this["val"])
        else:
    #         this["val"].values
    #         assert len(this)==1, "baseline should be a single item"
            baseval = this["val"].item()
        return baseval
    dfdat_simple = aggregThenReassignToNewColumn(dfdat_simple, F, ["var", "chan", "load_suffix"], "val_baseline")
    dfdat = aggregThenReassignToNewColumn(dfdat, F, ["var", "chan", "load_suffix"], "val_baseline")

    # for each row subtrract baseline
    def F(x):
        return x["val"] - x["val_baseline"]
    dfdat_simple = applyFunctionToAllRows(dfdat_simple, F, "val_minus_baseline")
    dfdat = applyFunctionToAllRows(dfdat, F, "val_minus_baseline")
        
    return dfdat, dfdat_simple, SP
```
