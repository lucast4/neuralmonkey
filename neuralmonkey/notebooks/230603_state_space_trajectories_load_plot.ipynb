```python
######## SINGLE TRIAL
# extract a data snippet to plot
def _extract_frmat(index_datapt, event, output_dim=2):
    
    dfthis = SP.dataextract_as_df_good(event=event, var="index_datapt", var_level=index_datapt)

    # dfthis_to_frmat(dfthis, output_dim=2)
    assert len(dfthis)>0
    
    # make sure chans match the desired chans
    if len(dfthis["chan"])<len(PApca.Chans):
        # Not enough chans
#         print(dfthis["chan"])
#         print(len(PApca.Chans))
        return None
    elif len(dfthis["chan"])>len(PApca.Chans):
        if all([c in dfthis["chan"].tolist() for c in PApca.Chans]):
            # Too many chans, then just take just the chans desired 
            from pythonlib.tools.pandastools import slice_by_row_label
            dfthis = slice_by_row_label(dfthis, "chan", PApca.Chans, reset_index=True, assert_exactly_one_each=True)
        else:
            # too many chans, and doesnt have all required chans
            return None
    else: # same length
        if not np.all(dfthis["chan"] == PApca.Chans):
            # mismatched chans...
            return None
        else:
            # Good.
            pass
        
    # get frmat
    frmat, _ = dfthis_to_frmat(dfthis, None, which_fr_sm, output_dim=output_dim, time_bin_size=0.02)
    return frmat


######## AGG OVER TRIALS
from pythonlib.tools.pandastools import aggregThenReassignToNewColumn, aggregGeneral

def _extract_frmat_aggtrials(list_index_datapt, event):
    # collect each trial, then concat the frmats

    # dfthis = SP.dataextract_as_df_good(event=event)
    # # agg across trials
    # aggregGeneral(dfthis, group=["chan"], values=)

    list_frmat = []
    for index_datapt in list_index_datapt:
        frmat = _extract_frmat(index_datapt, event, output_dim=3)

        if frmat is None:
            print(f"Skipping index_data: {index_datapt}, since it doesnt have all chans..)")
            continue
    #     print(frmat.shape)

        list_frmat.append(frmat)

    frmat_alldat = np.concatenate(list_frmat, axis=1)
    frmat = np.mean(frmat_alldat, axis=1)
    
    return frmat

list_index_datapt = SP.DfScalar["index_datapt"].unique()
```
