```python
# plot looking before/during + drawing beh for a GIVEN SHAPE (see above cell)

# given events, plot eye-tracking beh
def plotLookingBehaviorBetweenEvents(sn, ax, trial, start_event, end_event):
    title = "looking behavior between: " + start_event + " and " + end_event
    ax.set_title(title)

    # 1) plot task image
    sn.plot_taskimage(ax, trial)

    # 2) get eye tracking data (smooothed)
    times_tdt, xy_tdt_calib, fs_tdt, xy_tdt_calib_sm = sn.beh_extract_eye_good(trial, PLOT=False, return_all=True);

    # keep just times between [start_event, end_event]
    dict_event_times = sn.events_get_time_sorted(trial, list_events=(start_event, end_event))[0]
    start_time = dict_event_times[start_event]
    end_time = dict_event_times[end_event]
    
    event_inds = (times_tdt>=start_time) & (times_tdt<=end_time)
    event_times = times_tdt[event_inds]
    
    event_xy_vals = xy_tdt_calib_sm[event_inds] # 2 columns, x and y
    event_x_vals = event_xy_vals[:,0]
    event_y_vals = event_xy_vals[:,1]

    # 3) plot eye trace over task image
    plotEyeTrace(event_times, event_x_vals, event_y_vals, ax)
    
    # 4) plot drawing behavior over task image
    #sn.Datasetbeh.plot_strokes([event_xy_vals], ax, single_color="r")

#### testing ####
#trials_list = sn.get_trials_list(True)
#n=10
#trials = random.sample(trials_list, n)
#print("trials", trials)
#fig, axes = plt.subplots(n, 3, figsize=(30,30))
trials = s_all_trials
n = len(trials)
fig, axes = plt.subplots(n, 3, figsize=(30, 30))

# make plot grid
for i in range(len(trials)):
    t = trials[i]
    ax = axes[i,0]
    
    # plot eye-tracking beh BEFORE drawing
    plotLookingBehaviorBetweenEvents(sn, axes[i,0], t, "stim_onset", "go")
    
    # plot drawing beh
    sn.plot_taskimage(axes[i,1], t)
    plotAllDrawnStrokes(sn, t, axes[i,1])
    
    # plot eye-tracking beh DURING drawing
    plotLookingBehaviorBetweenEvents(sn, axes[i,2], t, "go", "off_stroke_last")
    
#### OBSERVATIONS ####
# - unclear whether he is tracing shape with eyes
# - seems to fixate on centroid of staple
```
